```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from statsmodels.tsa.api import seasonal_decompose
```

```python
pasajeros = pd.read_csv('airline_passengers.csv',index_col='Month',parse_dates=True)
pasajeros
```

```python
pasajeros = pasajeros.rename(columns={'Thousands of Passengers':'Miles de pasajeros'},)
```

```python
pasajeros.plot(figsize=(12,6))
```

```python
Serie1 = seasonal_decompose(pasajeros['Miles de pasajeros'],model='mul')
Serie1.plot()
```

```python
pasajeros['serie desestacionalizada']=pasajeros['Miles de pasajeros']/Serie1.seasonal
```

```python
pasajeros['serie desestacionalizada'].plot()
```
