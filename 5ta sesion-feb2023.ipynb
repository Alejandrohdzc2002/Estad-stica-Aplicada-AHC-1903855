```python
plt.bar(mun_nl['municipio'],mun_nl['nacimientos'],color=['blue','green','red','yellow'])
plt.title('Nacimientos por municipio') #Asi se agrega el titulo de gráfico
plt.xlabel('Municipios de NL') #Nombre Eje X
plt.ylabel('Cantidad de nacimientos') #Nombre Eje y
plt.yticks([0,200,400,600,800])#Así se agrega la escala de el eje y
plt.show()
```

```python
#Ordernar data frame
mun_nl=mun_nl.sort_values('nacimientos',ascending = False)
```
