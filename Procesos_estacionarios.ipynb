```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
#Las series de tiempo
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import adfuller
%matplotlib inline
```

```python
procesos = pd.read_excel('Procesos estacionarios.xlsx')
procesos
```

```python
# Gráfica de la serie de tiempo
procesos['ruido_blanco'].plot(color='red',linewidth=.4)
plt.title('Proceso de ruido blanco')
plt.xlabel('tiempo')
plt.xlim(0,1000)
```

1. **Comprobar si el proceso es estacionario**\
    Esto se comprueba con la serie de Fuller

```python
adfuller(procesos['ruido_blanco'])
# nivel de significancia del 5%

alpha = .5
if adfuller(procesos['ruido_blanco'])[1] < alpha:
    print('Se rechaza Ho, la serie es estacionaria')
else:
    print('No se rechaza Ho, la serie no es estacionaria')
```

2. **Verificar que tipo de proceso estacionario es**

```python
fig, ax=plt.subplots(1,2,figsize=(12,4))
## subgrafico 0
plot_acf(procesos['ruido_blanco'],zero=False,ax=ax[0],title='Función de autocorrelacion ACF')
ax[0].set_ylim(-1,1)
## subgrafico 1
plot_pacf(procesos['ruido_blanco'],zero=False,ax=ax[1],title='Función de autocorrelación parcial PACF')
ax[1].set_ylim(-1,1)
```
